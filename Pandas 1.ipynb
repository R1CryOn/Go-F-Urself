```python
import pandas as pd
sales = pd.read_csv('sales_subset.csv')
sales.head(2)
```

```python
sales.drop_duplicates('store').head()
```

```python
sales.drop_duplicates(subset=["store", "type"])
```

```python
# Subset the rows where is_holiday is True and drop duplicate dates
sales[sales["is_holiday"]].drop_duplicates(subset="date")
```

```python
sales['type'].value_counts()
```

```python
sales['type'].value_counts(normalize=True)
```

```python
sales['type'].value_counts(sort=True)
```

```python
sales.groupby("type")["weekly_sales"].mean()
```

```python
sales.groupby("type")["weekly_sales"].agg(['mean','median'])
```

```python
# Pivot for mean weekly_sales for each store type
sales.pivot_table(values="weekly_sales", index="type")
```

```python
import numpy as np
sales.pivot_table(values="weekly_sales", index="type", aggfunc=[np.mean, np.median])
```

```python
sales.pivot_table(values="weekly_sales", index="type", columns="is_holiday")
```
